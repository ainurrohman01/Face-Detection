# Static Face Detection (Image/Photo)

```python
# Face Detection Only

import cv2

# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Read the input image
img = cv2.imread('single face.jpg')

# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 6)  

# Display the output
cv2.imshow('img', img)

#Save gambar
cv2.imwrite('deteksi wajah single face.jpg', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Face & Eye Detection

import cv2

# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cascade_ = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses.xml')
# Cascade is Machine Learning approach (Traditional ML) Pre-Trained classifiers for faces, eyes, and smile use XML file.
""" Store :
– data/haarcascade_frontalface_default.xml
– data/haarcascade_eye.xml
– data/haarcascade_eye_tree_eyeglasses.xml

haarcascade_eye_tree_eyeglasses.xml is an eye detector with better handling of eyeglasses.
"""

# Read the input image
img = cv2.imread('multi face.jpg')

# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
faces_ = face_cascade_.detectMultiScale(gray, 1.1, 4)  # Untuk yang deteksi mata
# detecMultiScale() function is for detecting objects if it finds a face in the image it will return in the form of x,y,w,h. and it needs some parameters.
# 1.1 = ScaleFactor ==> This is parameter is for specifying  how much the image size is reduced at each image scale.
# 4   = minNeighbors ==> Parameter specifying how many neighbors each candidate rectangle should have to retain it, this parameter will affect the quality of the detected faces.

# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 255), 4)
    
    # # Untuk yang deteksi mata
for (x, y, w, h) in faces_:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)    


# Display the output
cv2.imshow('img', img)

#Save gambar
cv2.imwrite('deteksi wajah dan mata - multi face.jpg', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

# Dynamic Face Detection (Video)

```python
# Mendeteksi wajah dan mata dengan Webcam dan menyimpannya dalam bentuk video 

import cv2

# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cascade_ = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses.xml')

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc('X','V','I','D') # FourCC is a 4-byte code used to specify the video codec. The codecs for Windows is DIVX and for OSX is avc1, h26
videoWriter = cv2.VideoWriter('video.avi', fourcc, 30.0, (640,480))
# Referensi untuk save video: https:
""" (1). https://www.geeksforgeeks.org/saving-operated-video-from-a-webcam-using-opencv/
    (2). https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_gui/py_video_display/py_video_display.html
"""

cap = cv2.VideoCapture(0)
#help(cv2)
while cap.isOpened():
    #BGR image feed from camera
    ret, img = cap.read()
    cv2.imshow('Original Video', img)
    #BGR to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #cv2.imshow('grayscale', gray)
     
    
    faces = face_cascade.detectMultiScale(gray, 1.1, 5) #minSize=(30, 30)
    
    faces_ = face_cascade_.detectMultiScale(gray, 1.1, 5)
    
    # Draw rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 255), 4)
    
    # # Untuk yang deteksi mata
    for (x, y, w, h) in faces_:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)  

    # Display the resulting frame
    cv2.imshow('Output', img)

    # Save video
    videoWriter.write(img)
    
    k = cv2.waitKey(10)
    if k==27:
        break
        

cap.release()
videoWriter.release()
cv2.destroyAllWindows()

```
